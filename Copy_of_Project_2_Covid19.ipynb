```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/08-02-2020.csv')

df.dropna(inplace= True)
# Drop NAs

df.drop(['Admin2','Combined_Key','FIPS','Lat','Long_'],axis=1,inplace=True)
# Drop Columns

df[['Active']]=(df[['Active']]*1).astype(int)
# Drop Decimal from Active

df = df.round({'Incidence_Rate': 1, 'Case-Fatality_Ratio': 1})
# Round Incidence and Case Fatality

df[['Update','Time']] = df.Last_Update.str.split(" ",expand=True,)
# Split Update Column

df.drop(['Last_Update','Time'],axis=1,inplace=True)
# Drop Columnns

df=df.rename(columns={'Province_State':'State','Country_Region': 'Country', 'Confirmed': 'Cases','Case-Fatality_Ratio': 'Cases-Deaths Ratio', 'Incidence_Rate': 'Case Rate'})
# Rename Column

df=df.sort_values(['State'], ascending = True)
# Sort by State

df1 = df.groupby(['State'])['Cases','Deaths','Active'].sum()
# Create a new sheet grouped by State

df1=df1.sort_values(['Cases'], ascending=False)
# Sort sheet grouped by State by total number of cases

df1
```
